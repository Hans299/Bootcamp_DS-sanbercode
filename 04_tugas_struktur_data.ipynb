## Soal 1
**Mengolah Data dengan List dan Tuple**
Seorang guru memiliki daftar nilai siswa dalam bentuk list, seperti berikut:  
`nilai_siswa = [85, 90, 78, 92, 88, 76, 95, 89, 77, 84]`

#### **Tugas:**
1. **List:**
   - Hitung rata-rata nilai siswa.
   - Tentukan nilai tertinggi dan nilai terendah.
   - Urutkan nilai siswa dari yang terendah ke tertinggi.

2. **Tuple:**
   - Ubah list yang telah dibuat menjadi tuple.
   - Jelaskan mengapa tuple cocok untuk digunakan jika data tidak akan diubah.

#### **Output yang Diharapkan:**
```python
1. List:
   Rata-rata nilai: 85.4
   Nilai tertinggi: 95
   Nilai terendah: 76
   Daftar nilai terurut: [76, 77, 78, 84, 85, 88, 89, 90, 92, 95]

2. Tuple:
   Tuple nilai siswa: (85, 90, 78, 92, 88, 76, 95, 89, 77, 84)
   Nested list dan tuple: [[85, 90, 78, 92, 88, 76, 95, 89, 77, 84], (85, 90, 78, 92, 88, 76, 95, 89, 77, 84)]
```

---

````python
nilai_siswa = [85, 90, 78, 92, 88, 76, 95, 89, 77, 84]
banyak_nilai=len(nilai_siswa)
print('1')
#sort dengan sort()
print('#memakai sort()')
nilai_siswa.sort()
nilai_siswa.sort(reverse=True)
print('Nilai Yang sudah disort = ',nilai_siswa)
print('Nilai yang sudah disort dari terbesar',nilai_siswa)
#sort dengan bubblesort alogarthim
print('#sort dengan bubblesort alogarthim')
# Perulangan untuk proses pengurutan
for i in range (banyak_nilai):
    # Perulangan untuk proses perbandingan nilai
    for j in range (banyak_nilai-1):
        # Proses pengecekan nilai
        if(nilai_siswa[j]<nilai_siswa[j-1]):
            # Proses pertukaran nilai
            nilai_siswa[j],nilai_siswa[j-1]=nilai_siswa[j-1],nilai_siswa[j]
print('Nilai yang diurutkan dengan bubbleSort (kecil ke besar)= ',nilai_siswa)
for i in range (banyak_nilai):
    for j in range (banyak_nilai-1):
        if(nilai_siswa[j]>nilai_siswa[j-1]):
            nilai_siswa[j],nilai_siswa[j-1]=nilai_siswa[j-1],nilai_siswa[j]
print('Nilai yang diurutkan dengan bubbleSort (besar ke kecil) =  ',nilai_siswa)
total_nilai=0
for nilai in nilai_siswa:
    total_nilai+=nilai
print('Rata-rata Nilai = ',total_nilai/banyak_nilai)
#Max Min Value
nilai_sementara=nilai_siswa[0] #nilai sementara yang dideklarasikan untuk mulai cek
for i in range (banyak_nilai): # perulangan pengecekan nilai
    if(nilai_siswa[i]>nilai_sementara): #pengecekan nilai
        nilai_sementara=nilai_siswa[i] #pembaruan nilai
print('Nilai tertinggi = ',nilai_sementara)
for i in range (banyak_nilai):
    if(nilai_siswa[i]<nilai_sementara):
        nilai_sementara=nilai_siswa[i]
print('Nilai tertinggi = ',nilai_sementara)
print('2')
nilai_tuple=tuple(nilai_siswa)
print('nilai tuple = ',nilai_tuple)
print('Tuple cocok digunakan jika data tidak akan diubah karena tuple bersifat immutable (tidak bisa diubah setelah dibuat), sehingga data lebih aman dan stabil.')
````
