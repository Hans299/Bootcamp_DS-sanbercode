## SOAL 1
**Analisis Ukuran Pemusatan Data Tinggi Badan Siswa**
Data tinggi badan (dalam cm) dari 15 siswa di sebuah kelas adalah:

160, 165, 170, 175, 180, 165, 170, 175, 180, 160, 170, 175, 165, 160, 180

Lakukan analisis berikut:

1.  Hitung mean (rata-rata) dari data tinggi badan tersebut.
2.  Hitung median (nilai tengah) dari data tinggi badan tersebut.
3.  Tentukan modus (nilai yang paling sering muncul) dari data tinggi badan tersebut.
4.  Interpretasikan apa yang digambarkan oleh ukuran pemusatan data ini tentang distribusi tinggi badan di kelas tersebut.

## JAWABAN SOAL 1


count (15.000000): Jumlah total data observasi. Ada 15 buah data dalam kumpulan ini.

mean (170.000000): Rata-rata hitung (average). Jika semua nilai dijumlahkan lalu dibagi 15, hasilnya adalah 170.

std (7.319251): Standar Deviasi atau simpangan baku. Ini mengukur sebaran data. Nilai sekitar 7.32 menunjukkan seberapa jauh, rata-rata, nilai-nilai tersebut menyimpang dari angka rata-ratanya (170).

min (160.000000): Nilai terkecil (minimum) di dalam data tersebut adalah 160.

25% (165.000000): Kuartil pertama (Q1). Artinya, 25% dari data bernilai 165 atau lebih kecil.

50% (170.000000): Median atau nilai tengah (Q2). Jika data diurutkan dari terkecil ke terbesar, nilai yang persis di tengah adalah 170. Perhatikan bahwa nilai median sama dengan nilai rata-rata (mean), yang biasanya menandakan data berdistribusi simetris.

75% (175.000000): Kuartil ketiga (Q3). Artinya, 75% dari data bernilai 175 atau lebih kecil.

max (180.000000): Nilai terbesar (maksimum) di dalam data tersebut adalah 180.

dtype: float64: Tipe data yang digunakan komputer untuk menyimpan angka ini adalah angka desimal (floating-point 64-bit).

Nilai yang sering muncul (modus) = [160 165 170 175 180]: Data ini memiliki lebih dari satu modus (multimodal). Angka-angka tersebut muncul dengan frekuensi yang sama tingginya.

```python
import pandas as pd
import numpy as np
tinggi_badan=[160, 165, 170, 175, 180, 165, 170, 175, 180, 160, 170, 175, 165, 160, 180]
tinggi_badan.sort()
seri_tinggi=pd.Series(tinggi_badan) #sebelum memakai fungsi module maka buat dulu data menjadi komponen sesuai syarat module
arr_tinggi=np.array(tinggi_badan) #cara pakai module dan inisial.komponen data
print('Ringkasan Data')
print(seri_tinggi.describe())
print('Nilai yang sering muncul (modus) = ', np.unique(arr_tinggi))
```

## SOAL 2
**Menghitung Ukuran Penyebaran Data**

Data berat badan (dalam kg) dari 10 siswa adalah:

50, 55, 60, 65, 70, 55, 60, 65, 70, 75

Lakukan analisis berikut:

1.  Hitung *range* (rentang) dari data berat badan.
2.  Hitung varians (ragam) dari data berat badan.
3.  Hitung standar deviasi (simpangan baku) dari data berat badan.
4.  Jelaskan seberapa besar penyebaran data berat badan ini dari rata-rata (mean).

## JAWABAN SOAL 2
Jika kita menggabungkan Mean (62.5) dan Standar Deviasi (7.5), kita bisa melihat pola penyebaran data Anda dengan sangat jelas:

Batas Bawah 1 Standar Deviasi: 62.5 - 7.5 = 55

Batas Atas 1 Standar Deviasi: 62.5 + 7.5 = 70

Jika Anda perhatikan kembali kumpulan data Anda (50, 55, 60, 65, 70, 55, 60, 65, 70, 75), mayoritas orang (8 dari 10 orang) memiliki berat badan yang berada persis di dalam rentang penyebaran normal ini (antara 55 kg hingga 70 kg). Data menyebar secara sangat rapi dan simetris di kiri dan kanan angka 62.5.

```python
import numpy as np
berat_badan = [50, 55, 60, 65, 70, 55, 60, 65, 70, 75]
berat_badan.sort()
arr_data=np.array(berat_badan) #sebelum memakai fungsi module maka buat dulu data menjadi komponen sesuai syarat module
print('max value = ', max(arr_data))
print('min value = ',min(arr_data))
print('Range = ',max(arr_data)-min(arr_data))
print('variasi (ragam) = ',np.var(arr_data)) #cara pakai module dan inisial.komponen data
print('standar deviasi = ',np.std(arr_data)) #cara pakai module dan inisial.komponen data
print('Mean = ',np.mean(arr_data)) #cara pakai module dan inisial.komponen data
```

## SOAL 3
**Kombinasi Pemusatan dan Penyebaran Data**

Hasil ujian matematika 12 siswa adalah:

72, 75, 78, 80, 85, 90, 92, 95, 98, 100, 105, 110

Lakukan analisis berikut:

1.  Hitung mean (rata-rata), median (nilai tengah), dan modus (nilai yang paling sering muncul).
2.  Hitung range (rentang), variance (ragam), dan standard deviation (simpangan baku).
3.  Buat kesimpulan tentang performa siswa berdasarkan ukuran pemusatan dan penyebaran data.

## JAWABAN SOAL 3
Secara kolektif, performa belajar kelompok siswa ini tergolong sangat baik dan solid (ditandai dengan rata-rata dan median di angka 90-an). Mayoritas siswa terbukti mampu menyerap materi dengan sangat optimal.

Namun, dari sisi pemerataan, terdapat ketimpangan atau variasi kemampuan yang perlu diperhatikan (ditandai dengan standar deviasi 12.19 dan jangkauan 38 poin). Ada segelintir siswa yang sangat over-perform (mencapai skor 105 dan 110), sementara di sisi lain ada sekitar 3 siswa di kuartil bawah yang nilainya tertinggal cukup jauh (berkisar di angka 72 - 78) dibandingkan mayoritas teman sekelasnya.

```python
import pandas as pd
import numpy as np
nilai=[72, 75, 78, 80, 85, 90, 92, 95, 98, 100, 105, 110]
nilai.sort()
series_nilai=pd.Series(nilai) #sebelum memakai fungsi module maka buat dulu data menjadi komponen sesuai syarat module
arr_nilai=np.array(nilai)
print('Variansi = ',np.var(arr_nilai))
print('Modus = ',series_nilai.mode())
print('ringkasan data')
print(series_nilai.describe()) #cara pakai module dan inisial.komponen data
```
