### Soal 2 (Tipe: Kode)

Buatlah kode Python menggunakan library Pandas untuk melakukan langkah-langkah berikut:

1.  Import library pandas.
2.  Buatlah sebuah dictionary Python dengan nama `data` yang berisi data sebagai berikut:
    ```python
    {
        'Nama': ['Andi', 'Budi', 'Citra', 'Dewi', 'Eko', 'Fani'],
        'Usia': [25, 30, 22, 28, 35, 27],
        'Kota': ['Jakarta', 'Surabaya', 'Bandung', 'Yogyakarta', 'Semarang', 'Medan']
    }
    ```
3.  Konversikan dictionary `data` tersebut menjadi sebuah Pandas DataFrame.
4.  Tampilkan 3 baris pertama dari DataFrame menggunakan method `.head()`.
5.  Tampilkan ringkasan informasi mengenai DataFrame (jumlah non-null, tipe data, dll.) menggunakan method `.info()`.


````python
import pandas as pd
data = {
    'Nama': ['Andi', 'Budi', 'Citra', 'Dewi', 'Eko', 'Fani'],
    'Usia': [25, 30, 22, 28, 35, 27],
    'Kota': ['Jakarta', 'Surabaya', 'Bandung', 'Yogyakarta', 'Semarang', 'Medan']
}
df_data=pd.DataFrame(data)
print('frame data = \n',df_data) 
print('frame data = \n',df_data.head(3)) #cara pakai fungsi panda (komponen.fungsi)
display('info frame data = \n',df_data.info()) #alternatif print adalah display
````

### Soal 4 (Tipe: Kode)

Anda diberikan sebuah dictionary Python berikut yang berisi data siswa:

```python
data = {'nama': ['Andi', 'Budi', 'Citra', 'Dewi'],
        'usia': [20, 21, 22, 20],
        'nilai': [85, 92, 78, 88]}
```

1.  Buatlah sebuah Pandas DataFrame dari dictionary `data` tersebut.
2.  Pilih kolom dengan nama `nilai` dari DataFrame tersebut.
3.  Hitung nilai rata-rata (mean) dari kolom `nilai` menggunakan method atau fungsi yang sesuai dari Pandas atau NumPy.
4.  Cetak hasil nilai rata-rata tersebut.


````python
import pandas as pd
import numpy as np
data = {
    'nama': ['Andi', 'Budi', 'Citra', 'Dewi'],
    'usia': [20, 21, 22, 20],
    'nilai': [85, 92, 78, 88]
}
df_data=pd.DataFrame(data)
kolom_nilai=df_data['nilai']
arr_nilai=np.array(kolom_nilai)
print('Nilai = \n',kolom_nilai)
print('Rata-rata = ', np.mean(arr_nilai))
````

#### Expected Output Soal 4

```text
85.75
```

3.  Tampilkan informasi ringkasan mengenai DataFrame, termasuk tipe data dan jumlah non-null per kolom, menggunakan metode `.info()`.
4.  Tampilkan statistik deskriptif (seperti mean, min, max, standard deviation, dll.) untuk kolom numerik dalam DataFrame menggunakan metode `.describe()`.


````python
import pandas as pd
data = {
    'Nama': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Usia': [25, 30, 35, 28, 32],
    'Nilai': [85.5, 92.0, 78.5, 88.0, 95.5]
}
df_data=pd.DataFrame(data)
atas=df_data.head(3)
print('top 3 = \n',atas)
display('desc = \n',atas.describe())
display('info = \n',atas.info())


````
