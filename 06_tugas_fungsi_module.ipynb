### 
### import nama_modul (Ini mengimpor seluruh modul. Semua definisi (fungsi, variabel, dll.) di dalam modul tersebut dibawa ke dalam program) sementara from nama_modul import nama_fungsi_atau_variabel ( Ini hanya mengimpor item spesifik (fungsi, variabel, atau kelas) dari modul langsung ke namespace program Anda saat ini) dan from nama_modul import (Ini mengimpor semua nama (yang tidak diawali underscore _) dari modul langsung ke namespace Anda saat ini)

####
1. import nama_modul
Perintah ini digunakan untuk mengimpor seluruh modul sehingga semua fungsi, variabel, atau kelas yang ada di dalam modul tersebut dapat digunakan dalam program. Namun, untuk mengaksesnya harus menggunakan nama modul sebagai prefix. Misalnya jika menggunakan import math, maka fungsi akar dipanggil dengan math.sqrt().Keuntungan dari cara ini adalah lebih aman dari konflik penamaan, karena semua elemen tetap berada di dalam namespace modul. Kekurangannya, penulisan kode menjadi lebih panjang karena setiap pemanggilan fungsi harus diawali dengan nama modul.

2. from nama_modul import nama_fungsi_atau_variabel
Perintah ini digunakan untuk mengimpor item tertentu saja dari sebuah modul, misalnya fungsi atau variabel tertentu. Item yang diimpor akan langsung masuk ke namespace program, sehingga dapat dipanggil tanpa menuliskan nama modul.
Keuntungannya adalah kode menjadi lebih singkat dan mudah dibaca, karena tidak perlu menuliskan nama modul setiap kali memanggil fungsi. Namun, kekurangannya adalah berpotensi terjadi konflik penamaan jika terdapat fungsi atau variabel lain dengan nama yang sama di dalam program atau dari modul lain.

3. from nama_modul import *
Perintah ini digunakan untuk mengimpor semua nama dari modul (kecuali yang diawali dengan underscore _) langsung ke namespace program. Dengan cara ini, semua fungsi dan variabel dari modul dapat digunakan secara langsung tanpa menuliskan nama modul.Keuntungannya adalah memudahkan penggunaan banyak fungsi sekaligus karena semuanya langsung tersedia. Namun, cara ini memiliki kekurangan yaitu risiko konflik penamaan sangat besar, karena semua nama dari modul masuk ke program dan bisa bertabrakan dengan variabel atau fungsi lain. Selain itu, kode juga menjadi lebih sulit dibaca dan dipahami karena tidak jelas fungsi tersebut berasal dari modul mana.
