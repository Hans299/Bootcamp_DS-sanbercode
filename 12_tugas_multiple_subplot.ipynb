### Soal 1

Buatlah kode Python menggunakan library `matplotlib.pyplot` untuk menghasilkan sebuah figure yang berisi grid 2x2 (dua baris, dua kolom) subplots.

- Gunakan fungsi `plt.figure()` untuk membuat figure dan mengatur ukurannya menjadi `(8, 6)` inci.
- Gunakan fungsi `plt.subplot()` dengan parameter yang sesuai (jumlah baris, jumlah kolom, indeks subplot) untuk memilih dan menggambar di setiap sel grid.
- Di setiap subplot (indeks 1, 2, 3, dan 4), gambarkan sebuah grafik garis sederhana (misalnya, garis lurus, gelombang sinus, gelombang kosinus, atau kurva eksponensial).
- Berikan judul yang sesuai untuk setiap subplot.
- Tambahkan judul keseluruhan figure.
- Sesuaikan jarak antar subplot secara otomatis.
- Tampilkan figure yang telah dibuat.

```python
import matplotlib.pyplot as plt
fig, ax =plt.subplots(2,2, figsize=(8,6))

ax[0,0].set_title('Garis Lurus')
ax[0,1].set_title('Gelombang Sinus')
ax[1,0].set_title('Gelombang Kosinus')
ax[1,1].set_title('Gelombang eksponesial')
plt.show()
```

### Soal 2

Gunakan pustaka Matplotlib di Python untuk membuat sebuah figure yang menampilkan 4 subplot dalam tata letak grid 2x2. Masing-masing subplot harus memvisualisasikan fungsi matematika yang berbeda dalam rentang x dari -10 hingga 10.

Detail untuk setiap subplot:

1.  **Kiri Atas (Row 0, Col 0):** Plot fungsi $y = \sin(x)$. Gunakan warna biru untuk garis dan beri judul "Sin(x)".
2.  **Kanan Atas (Row 0, Col 1):** Plot fungsi $y = \cos(x)$. Gunakan warna merah untuk garis dan beri judul "Cos(x)".
3.  **Kiri Bawah (Row 1, Col 0):** Plot fungsi $y = x$. Gunakan warna hijau untuk garis dan beri judul "y=x".
4.  **Kanan Bawah (Row 1, Col 1):** Plot fungsi $y = x^2$. Gunakan warna ungu untuk garis dan beri judul "y=x^2".

Persyaratan tambahan:

*   Setiap subplot harus memiliki label sumbu x ('X') dan sumbu y ('Y').
*   Atur ukuran keseluruhan figure menjadi 10x8 inci (`figsize`).
*   Tambahkan judul utama untuk seluruh figure.
*   Atur spasi antar subplots secara otomatis demi tampilan yang rapi.

```python
import matplotlib.pyplot as plt
fig, ax =plt.subplots(2,2, figsize=(10,8))

ax[0,0].set_title('sin(x)')
ax[0,1].set_title('cos(x)')
ax[1,0].set_title('y=x')
ax[1,1].set_title('y=x^2')
plt.show()
```
