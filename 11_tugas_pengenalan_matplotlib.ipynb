### Soal 1

Impor library `matplotlib.pyplot` dan buat grafik garis sederhana (line plot) dari data berikut:

```
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

Pastikan grafik ditampilkan.

````python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,5,7,11]
plt.plot(x,y)
plt.show()
````

### Soal 2

Buatlah kode Python menggunakan library Matplotlib untuk menghasilkan sebuah *bar chart* (diagram batang). Diagram tersebut harus menampilkan data penjualan untuk tiga produk berbeda. Data yang digunakan adalah sebagai berikut:

- Produk A: 150
- Produk B: 220
- Produk C: 180

Diagram batang tersebut harus memenuhi kriteria berikut:

1.  Memiliki judul: "Penjualan Produk Bulanan".
2.  Sumbu X berlabel: "Produk".
3.  Sumbu Y berlabel: "Jumlah Penjualan".
4.  Diagram disimpan dalam file bernama `penjualan_produk.png`.

````python
import matplotlib.pyplot as plt #bar pake .pyplot
produk = ['produk A','produk B','produk C']
nilai =[150,220,180]
plt.bar(produk,nilai)
plt.title('Jumlah Penjualan')
plt.xlabel('Produk')
plt.ylabel('Jumlah Penjualan')
plt.show()
````
