## Soal 1
**Sistem Penilaian Otomatis dengan Pengetesan Kondisi**

Buat sebuah script Python yang meminta pengguna memasukkan nilai siswa (dalam rentang 0-100). Program harus:
1. Menentukan dan mencetak grade berdasarkan kriteria berikut:
   - Nilai >= 90: Grade A
   - Nilai >= 80: Grade B
   - Nilai >= 70: Grade C
   - Nilai >= 60: Grade D
   - Nilai < 60: Grade E
2. Jika nilai yang dimasukkan tidak valid (di luar rentang 0-100), tampilkan pesan error.

**Pengetesan Kondisi:**
- Input: `90`
  Output:
  ```
  Grade: A
  ```
- Input: `65`
  Output:
  ```
  Grade: D
  ```
- Input: `-10`
  Output:
  ```
  Nilai tidak valid. Harus dalam rentang 0-100.
  ```

---

````python
nilai=int(input("Masukan Nilai : "))
if(nilai>100 or nilai<0):
    print('Nilai tidak valid. Harus dalam rentang 0-100.')
elif(nilai>=90):
    print('Grade A')
elif(nilai<90 and nilai>=80):
    print('Grade B')
elif(nilai<80 and nilai>=70):
    print('Grade C')
elif(nilai<70 and nilai>=60):
    print('Grade:D')
else:
    print('Grade E')
````
