## Soal 1
**Sistem Penilaian Otomatis dengan Pengetesan Kondisi**

Buat sebuah script Python yang meminta pengguna memasukkan nilai siswa (dalam rentang 0-100). Program harus:
1. Menentukan dan mencetak grade berdasarkan kriteria berikut:
   - Nilai >= 90: Grade A
   - Nilai >= 80: Grade B
   - Nilai >= 70: Grade C
   - Nilai >= 60: Grade D
   - Nilai < 60: Grade E
2. Jika nilai yang dimasukkan tidak valid (di luar rentang 0-100), tampilkan pesan error.

**Pengetesan Kondisi:**
- Input: `90`
  Output:
  ```
  Grade: A
  ```
- Input: `65`
  Output:
  ```
  Grade: D
  ```
- Input: `-10`
  Output:
  ```
  Nilai tidak valid. Harus dalam rentang 0-100.
  ```

---

````python
nilai=int(input("Masukan Nilai : "))
if(nilai>100 or nilai<0):
    print('Nilai tidak valid. Harus dalam rentang 0-100.')
elif(nilai>=90):
    print('Grade A')
elif(nilai<90 and nilai>=80):
    print('Grade B')
elif(nilai<80 and nilai>=70):
    print('Grade C')
elif(nilai<70 and nilai>=60):
    print('Grade:D')
else:
    print('Grade E')
````

### Soal 2
**Hitung Jumlah Huruf Vokal dan Konsonan dengan Pengetesan Kondisi**

Buat script Python untuk menghitung jumlah huruf vokal dan konsonan dalam sebuah kalimat yang dimasukkan pengguna. Abaikan karakter selain huruf. 

**Pengetesan Kondisi:**
1. Input: `"Halo Dunia"`
   Output:
   ```
   Jumlah huruf vokal: 5
   Jumlah huruf konsonan: 4
   ```
2. Input: `"123 AI!!"`
   Output:
   ```
   Jumlah huruf vokal: 2
   Jumlah huruf konsonan: 1
   ```
3. Input: `"Python"`
   Output:
   ```
   Jumlah huruf vokal: 1
   Jumlah huruf konsonan: 5
   ```

---

````python
konsonan=['b','B','c','C','d','D','f','F','g','G','h','H','j','J','k','K','l','L','m','M','n','N','p','P','q','Q','r','R','s','S','t','T','v','V','w','W','x','X','y','Y','z','Z']
vokal=['a','A','i','I','u','U','e','E','o','O']
count_konsonan=0
count_vokal=0
hitung_huruf=input('Input Kata = ')
for i in range (len(hitung_huruf)):
    if (hitung_huruf[i] in konsonan):
        count_konsonan+=1
    elif(hitung_huruf[i] in vokal):
        count_vokal+=1
print('Jumlah huruf vokal: ',count_vokal)
print('Jumlah huruf Konsonan: ',count_konsonan)
````

## Soal 3
**Simulasi Antrian dengan Pengetesan Kondisi**

Simulasikan antrian sederhana menggunakan script Python. script ini harus menerima input dari pengguna untuk menambah nama ke antrian, menampilkan daftar antrian, atau menghapus nama pertama dari antrian. 

**Pengetesan Kondisi:**
1. Tambahkan pelanggan: Budi dan Siti:
   ```python
   Menu:
   1. Tambahkan pelanggan ke antrian
   2. Tampilkan daftar antrian
   3. Layani pelanggan
   4. Keluar
   Pilih menu (1/2/3/4): 1
   Masukkan nama pelanggan: Budi
   Budi telah ditambahkan ke antrian.

   Pilih menu (1/2/3/4): 1
   Masukkan nama pelanggan: Siti
   Siti telah ditambahkan ke antrian.
   ```
2. Tampilkan daftar antrian:
   ```python
   Pilih menu (1/2/3/4): 2
   Daftar antrian: ['Budi', 'Siti']
   ```
3. Layani pelanggan
   ```python
   Pilih menu (1/2/3/4): 3
   Pelanggan yang dilayani: Budi
   ```
4. Tambahkan pelanggan: Ali dan tampilkan antrian
   ```python
   Pilih menu (1/2/3/4): 1
   Masukkan nama pelanggan: Ali
   Ali telah ditambahkan ke antrian.

   Pilih menu (1/2/3/4): 2
   Daftar antrian: ['Siti', 'Ali']
   ```
---

````python
print('Menu :')
print('1. Tambahkan pelanggan ke antrian')
print('2. Tampilkan daftar antrian')
print('3. Layani pelanggan')
print('4. keluar')

antrian = []

pilih = int(input('Pilih menu (1/2/3/4) : '))

while pilih == 1 or pilih == 2 or pilih == 3:

    if pilih == 1:
        nama = input('Masukkan nama pelanggan : ')
        antrian.append(nama)
        print(nama, 'telah ditambahkan ke antrian')

    elif pilih == 2:
        for pelanggan in antrian:
            print('Daftar antrian:', pelanggan)

    elif pilih == 3:
        if len(antrian) > 0:
            dilayani = antrian.pop(0)
            print('Pelanggan yang dilayani:', dilayani)
        else:
            print('Antrian kosong')

    pilih = int(input('Pilih menu (1/2/3/4) : '))
````
